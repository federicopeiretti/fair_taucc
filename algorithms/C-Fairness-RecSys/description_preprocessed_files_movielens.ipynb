```python
dict_users_map = {int(k.decode('utf-8')): v for k, v in data2["users_map"].items()}
dict_movies_map = {int(k.decode('utf-8')): v for k, v in data2["items_map"].items()}
```

```python
mapping_users = {
    "id_user": [],
    "id_row": []
}

for key, value in data2["users_map"].items():
    mapping_users["id_user"].append(int(key.decode('utf-8')))
    mapping_users["id_row"].append(int(value))
```

```python
pd.DataFrame(mapping_users).to_csv(root + "/movielens-1M_mapping_users.csv", index=False)
```

```python
mapping_movies = {
    "id_movie": [],
    "id_col": []
}

for key, value in data2["items_map"].items():
    mapping_movies["id_movie"].append(int(key.decode('utf-8')))
    mapping_movies["id_col"].append(int(value))
```

```python
pd.DataFrame(mapping_movies).to_csv(root + "/movielens-1M_mapping_movies.csv", index=False)
```
