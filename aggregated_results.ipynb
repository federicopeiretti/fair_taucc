```python
var_fair = df_fair.groupby(["fair_majority", "fair_minority1", "fair_minority2"]).var()
#var_fair = df_fair.groupby(["fair_majority", "fair_minority"]).var()
var_fair
```

```python
aggregated = pd.DataFrame()
aggregated
```

```python
for key in mean_fair.keys():
    aggregated[f"{key}_mean"] = mean_fair[key].values

for key in std_fair.keys():
    aggregated[f"{key}_std"] = std_fair[key].values
    
for key in var_fair.keys():
    aggregated[f"{key}_var"] = var_fair[key].values
```

```python
aggregated
```

```python
aggregated.insert(0, "fair_major", fair_major)
#aggregated.insert(1, "fair_minor", fair_minor1)
aggregated.insert(1, "fair_minor1", fair_minor1)
aggregated.insert(2, "fair_minor2", fair_minor2)
```

# Vanilla TauCC

```python
df_res = pd.read_csv(path_vanilla + "/results_runs.csv", sep=";")
df_res
```

```python
cols_metrics = df_res.columns.tolist()[4:]
cols_metrics
```

```python
stats_vanilla = {}

for metric in cols_metrics:
    values = df_res[metric].tolist()
    stats_vanilla[f"{metric}_mean"] = [np.mean(values)]
    stats_vanilla[f"{metric}_std"] = [np.std(values)]
    stats_vanilla[f"{metric}_var"] = [np.var(values)]
```

```python
stats_vanilla
```

```python
df_stats_vanilla = pd.DataFrame(stats_vanilla)
df_stats_vanilla.to_csv(path_vanilla + "/results_aggregated.csv", index=False)
```

## Best run (max tau_x)

```python
best_run = np.argmax(df_res["tau_x"])
print(f"Best run: {best_run}")
```

```python
pd.DataFrame(df_res[df_res["run"] == best_run]).to_csv(path_vanilla + "/best_run.csv", index=False)
```
